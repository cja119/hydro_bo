```python
import h3
import geopandas as gpd
from shapely.geometry import Polygon
import sys
from numpy import datetime64
import meteor_py
```

```python


start_date =  datetime64('2022-01-01', 'ns') 
end_date    = datetime64('2023-01-01', 'ns') 

weatherdata = meteor_py.WeatherData(
    date = (start_date,end_date),
    location= 'North Sea', 
    wind = True,
    solar = False, 
    latitudes =(53.5, 55.5), 
    longitudes = (0.5, 4.5)
    )


meteor_py.RenewableEnergy(
    weatherdata,
    [
        (0, 0.0),       # These are points along the power curve.
        (3, 0.0),       # are used in the output curve.
        (4, 0.648),     # Wind speeds are in [m/s].
        (5, 1.4832),
        (6, 2.736),
        (7, 4.4676),
        (8, 6.7104),
        (9, 9.3168),
        (10, 11.2392),
        (11, 11.8008),
        (12, 11.8728),
        (13, 11.88),
        (30, 11.88),
    ]
).export_power(
    weatherdata,
    name='north_sea_wind',
    dates=True
    )
```
